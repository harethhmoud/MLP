```python
import numpy as np
import pandas as pd

class NaiveBayes:

    def __init__(self, alpha=1):
        self.alpha = alpha 

    def fit(self, x, y):
        N, D = x.shape 
        C = np.max(y) + 1 

        Nc = np.zeros(C) 
        Fc = np.zeros((C, D)) 

        for i in range(N):
            c = y[i]
            Nc[c] += 1
            Fc[c] += x[i].toarray().squeeze()

        self.class_log_prior_ = np.log((Nc + self.alpha) / (N + C * self.alpha))
        self.feature_log_prob_ = np.log((Fc + self.alpha) / (Nc[:, np.newaxis] + D * self.alpha))

        return self

    def predict(self, xt):
        N, D = xt.shape 
        C = self.feature_log_prob_.shape[0] 

        log_likelihood = np.zeros((N, C))

        for i, row in enumerate(xt):
            row_array = row.toarray()
            for index in range(D):
                value = row_array[0, index]
                if value != 0:
                    log_likelihood[i] += value * self.feature_log_prob_[:, index]
            log_likelihood[i] += self.class_log_prior_

        y_pred = np.argmax(log_likelihood, axis=1)

        return y_pred

    def evaluate_acc(self, y_true, y_pred):
        correct_predictions = np.sum(y_true == y_pred)
        accuracy = correct_predictions / len(y_true)
        return accuracy*100

labels = np.zeros(len(cleaned_data),dtype = int)
labels[len(os.listdir(negative_folder_path)):] = 1

nb = NaiveBayes()

nb.fit(X, labels)

test_labels = np.zeros(len(cleaned_test_data),dtype=int)
test_labels[len(os.listdir(negative_test_folder_path)):] = 1

y_pred_test = nb.predict(X_test)

test_accuracy = nb.evaluate_acc(test_labels, y_pred_test)
print("Test Accuracy:", test_accuracy)
```

```python
import os
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
import html
import re
import warnings
import string
import tensorflow as tf
import sklearn
from tqdm import tqdm
from transformers import BertTokenizer, TFBertForSequenceClassification
from transformers import InputExample, InputFeatures

nltk.download('punkt',quiet = True)

negative_folder_path = "/content/aclImdb/train/neg"
positive_folder_path = "/content/aclImdb/train/pos"

data = []

for filename in os.listdir(negative_folder_path):
    with open(os.path.join(negative_folder_path, filename), "r", encoding="utf-8") as f:
        review = f.read()
    #data.append((review,0))
    data.append(review)

for filename in os.listdir(positive_folder_path):
    with open(os.path.join(positive_folder_path, filename), "r", encoding="utf-8") as f:
        review = f.read()
    #data.append((review,0))
    data.append(review)

#print(len(data))
#df = pd.DataFrame(data, columns=["review","label"])
# df = pd.DataFrame(data, columns=["review"])

nltk.download("stopwords", quiet = True)
stop_words = set(stopwords.words("english"))

cleaned_data = []
for review in data:
    review = html.unescape(review)
    review = re.sub(r"<.*?>", "", review)  # remove HTML tags
    review = review.lower()  # convert to lowercase
    review = re.sub(r"[^\w\s]", "", review)  # remove punctuation
    review = re.sub(r"\s+", " ", review)  # remove extra spaces
    cleaned_data.append(review)

print(cleaned_data[0])

warnings.filterwarnings("ignore", category=UserWarning, module='sklearn')

vectorizer = CountVectorizer(lowercase=True, stop_words='english', tokenizer=word_tokenize, strip_accents='unicode')
vectorizer.fit(cleaned_data)

vocabulary = list(vectorizer.vocabulary_.keys())

X = vectorizer.transform(cleaned_data)

negative_test_folder_path = "/content/aclImdb/test/neg"
positive_test_folder_path = "/content/aclImdb/test/pos"

test_data = []

for filename in os.listdir(negative_test_folder_path):
    with open(os.path.join(negative_test_folder_path, filename), "r", encoding="utf-8") as f:
        review = f.read()
    #data.append((review,0))
    test_data.append(review)

for filename in os.listdir(positive_test_folder_path):
    with open(os.path.join(positive_test_folder_path, filename), "r", encoding="utf-8") as f:
        review = f.read()
    #data.append((review,0))
    test_data.append(review)

nltk.download("stopwords", quiet = True)
stop_words = set(stopwords.words("english"))

cleaned_test_data = []
for review in test_data:
    review = html.unescape(review)
    review = re.sub(r"<.*?>", "", review)  # remove HTML tags
    review = review.lower()  # convert to lowercase
    review = re.sub(r"[^\w\s]", "", review)  # remove punctuation
    review = re.sub(r"\s+", " ", review)  # remove extra spaces
    cleaned_test_data.append(review)

print(cleaned_test_data[0])

warnings.filterwarnings("ignore", category=UserWarning, module='sklearn')

X_test = vectorizer.transform(cleaned_test_data)

vocabulary = list(vectorizer.vocabulary_.keys())

X_test = vectorizer.transform(cleaned_test_data)

model = TFBertForSequenceClassification.from_pretrained("bert-base-uncased")
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

labels = np.zeros(len(cleaned_data),dtype = int)
labels[len(os.listdir(negative_folder_path)):] = 1
test_labels = np.zeros(len(cleaned_test_data),dtype=int)
test_labels[len(os.listdir(negative_test_folder_path)):] = 1


def convert_data_to_examples(train, test, review, sentiment): 
    train_InputExamples = train.apply(lambda x: InputExample(guid=None, # Globally unique ID for bookkeeping, unused in this case
                                                          text_a = x[review], 
                                                          label = x[sentiment]), axis = 1)

    validation_InputExamples = test.apply(lambda x: InputExample(guid=None, # Globally unique ID for bookkeeping, unused in this case
                                                          text_a = x[review], 
                                                          label = x[sentiment]), axis = 1,)
  
    return train_InputExamples, validation_InputExamples

train_size = int(len(cleaned_data) * 0.3)  
train_indices = np.random.choice(np.arange(len(cleaned_data)), train_size, replace=False)

val_indices = [i for i in range(len(cleaned_data)) if i not in train_indices]

train_data = [(cleaned_data[i], labels[i]) for i in train_indices]

val_data = [(cleaned_data[i], labels[i]) for i in val_indices]

test_size = int(len(cleaned_test_data) * 0.3)
test_indices = np.random.choice(np.arange(len(cleaned_test_data)), test_size, replace=False)

test_data = [(cleaned_test_data[i], test_labels[i]) for i in test_indices]

np.random.shuffle(train_data)
np.random.shuffle(val_data)
np.random.shuffle(test_data)

train = pd.DataFrame({'review': [review for review, label in train_data], 'sentiment': [label for review, label in train_data]})
val = pd.DataFrame({'review': [review for review, label in val_data], 'sentiment': [label for review, label in val_data]})
test = pd.DataFrame({'review': [review for review, label in test_data], 'sentiment': [label for review, label in test_data]})


def convert_examples_to_tf_dataset(examples, tokenizer, max_length=128):
    features = [] # -> will hold InputFeatures to be converted later

    for e in tqdm(examples):
        input_dict = tokenizer.encode_plus(
            e.text_a,
            add_special_tokens=True,    # Add 'CLS' and 'SEP'
            max_length=max_length,    # truncates if len(s) > max_length
            return_token_type_ids=True,
            return_attention_mask=True,
            pad_to_max_length=True, # pads to the right by default # CHECK THIS for pad_to_max_length
            truncation=True
        )

        input_ids, token_type_ids, attention_mask = (input_dict["input_ids"],input_dict["token_type_ids"], input_dict['attention_mask'])
        features.append(InputFeatures( input_ids=input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids, label=e.label) )

    def gen():
        for f in features:
            yield (
                {
                    "input_ids": f.input_ids,
                    "attention_mask": f.attention_mask,
                    "token_type_ids": f.token_type_ids,
                },
                f.label,
            )

    return tf.data.Dataset.from_generator(
        gen,
        ({"input_ids": tf.int32, "attention_mask": tf.int32, "token_type_ids": tf.int32}, tf.int64),
        (
            {
                "input_ids": tf.TensorShape([None]),
                "attention_mask": tf.TensorShape([None]),
                "token_type_ids": tf.TensorShape([None]),
            },
            tf.TensorShape([]),
        ),
    )

train_InputExamples, validation_InputExamples = convert_data_to_examples(train,  val, 'review',  'sentiment')
train_data = convert_examples_to_tf_dataset(list(train_InputExamples), tokenizer)
train_data = train_data.shuffle(100).batch(32).repeat(2)

validation_data = convert_examples_to_tf_dataset(list(validation_InputExamples), tokenizer)
validation_data = validation_data.batch(32)

model.config.output_attentions=True
model.config.output_hidden_states=True

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=3e-5, epsilon=1e-08, clipnorm=1.0, weight_decay=0.5), 
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), 
              metrics=[tf.keras.metrics.SparseCategoricalAccuracy('accuracy')])

model.fit(train_data, epochs=1, validation_data=validation_data)
```
